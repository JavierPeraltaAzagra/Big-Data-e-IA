# Interpretación general de un gráfico 3D
En un gráfico 3D se muestran tres variables numéricas de un conjunto de datos al mismo tiempo.  
Cada punto representa una observación (una fila del DataFrame), y su color indica la clase a la que pertenece.

## Interpretación visual por clases

1. **Agrupamiento por color**  
   Si los puntos de una misma clase se agrupan en una región concreta del espacio 3D, significa que esas tres variables tienen una relación característica que distingue a esa clase.  
   Cuanto más compacto y separado esté ese grupo, más claramente la variable de clase se diferencia de las demás.

2. **Separación entre clases**  
   Si los grupos de colores están bien separados en el espacio tridimensional, las tres variables utilizadas son buenas para discriminar entre clases.  
   Si, en cambio, las nubes de puntos se superponen, las clases son más difíciles de distinguir usando solo esas variables.

3. **Forma de los grupos**  
   - Un grupo alargado puede indicar correlación entre dos de las variables.  
   - Un grupo más disperso sugiere que esa clase tiene mayor variabilidad en esas dimensiones.  
   - Si los grupos se cruzan o mezclan, es señal de que la frontera entre clases no es lineal o no puede definirse fácilmente.

4. **Densidad de puntos**  
   La concentración de puntos en una región indica dónde se acumulan los valores más frecuentes.  
   Regiones más vacías pueden corresponder a combinaciones poco comunes o imposibles de las variables.

## Conclusión

El gráfico 3D permite visualizar simultáneamente tres variables y analizar la relación entre ellas para diferentes clases.  
Si las clases aparecen bien separadas, esas variables podrían servir como buenos predictores en un modelo de clasificación.  
Si las clases se solapan o presentan fronteras difusas, puede ser necesario incluir más variables o aplicar técnicas no lineales para lograr una separación efectiva.

# Actividad: Visualización 3D con diferentes orientaciones

## Objetivo

Explorar la distribución de los datos de un conjunto (`DataFrame`) en tres dimensiones, usando tres variables numéricas.  
El objetivo es observar cómo **la orientación del gráfico 3D afecta la interpretación visual** de la separación entre clases o la relación entre variables.

## Descripción de la Actividad

1. Define una función que reciba como entrada:
   - Un `DataFrame`
   - Los nombres de **tres variables numéricas** (una por eje)
   - Dos valores numéricos (`elev` y `azim`) para definir la orientación

2. La función deberá generar un **gráfico 3D** con los datos del dataset, diferenciando las clases por color (asumiendo que la variable de clase se llama `'class'`).

3. Finalmente, llama a la función **seis veces** con diferentes orientaciones para mostrar seis vistas distintas del mismo gráfico.


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def grafico_3d(df, x_var, y_var, z_var, elev, azim):
    """
    Dibuja un gráfico 3D de tres variables numéricas con una orientación definida.
    Parámetros:
    df    : DataFrame con los datos
    x_var : nombre de la variable para el eje X
    y_var : nombre de la variable para el eje Y
    z_var : nombre de la variable para el eje Z
    elev  : ángulo de elevación de la cámara
    azim  : ángulo de rotación (azimut)
    """
    plt.figure(figsize=(8, 6))
    ax = plt.axes(projection='3d')
    plt.title(f'Gráfico 3D: {x_var}, {y_var}, {z_var} (elev={elev}, azim={azim})')
    # Clase 0
    x0 = df[df['class'] == 0][x_var]
    y0 = df[df['class'] == 0][y_var]
    z0 = df[df['class'] == 0][z_var]
    ax.scatter3D(x0, y0, z0, color='skyblue', alpha=0.7, label='Clase 0')
    # Clase 1
    x1 = df[df['class'] == 1][x_var]
    y1 = df[df['class'] == 1][y_var]
    z1 = df[df['class'] == 1][z_var]
    ax.scatter3D(x1, y1, z1, color='salmon', alpha=0.7, label='Clase 1')

    ax.set_xlabel(x_var)
    ax.set_ylabel(y_var)
    ax.set_zlabel(z_var)
    ax.legend()
    # Cambiar orientación
    ax.view_init(elev=elev, azim=azim)
    plt.tight_layout()
    plt.show()
```

```python
# Gráfico 1
grafico_3d(df, 'plas', 'mass', 'age', elev=20, azim=30)

# Gráfico 2
grafico_3d(df, 'plas', 'mass', 'age', elev=30, azim=60)

# Gráfico 3
grafico_3d(df, 'plas', 'mass', 'age', elev=45, azim=90)

# Gráfico 4
grafico_3d(df, 'plas', 'mass', 'age', elev=60, azim=120)

# Gráfico 5
grafico_3d(df, 'plas', 'mass', 'age', elev=15, azim=180)

# Gráfico 6
grafico_3d(df, 'plas', 'mass', 'age', elev=30, azim=270)
```

# Escalado de variables en el dataset Pima Indians Diabetes

## 1. Contexto del dataset

El conjunto de datos *Pima Indians Diabetes* contiene información médica de mujeres de origen Pima. Su objetivo es predecir si una persona tiene o no diabetes (variable `class` = 0 o 1) a partir de ocho variables numéricas:

| Variable | Descripción | Tipo de dato |
|-----------|--------------|--------------|
| preg | Número de embarazos | Entero |
| plas | Glucosa en plasma | Numérica continua |
| pres | Presión arterial diastólica | Numérica continua |
| skin | Grosor del pliegue cutáneo tricipital | Numérica continua |
| insu | Insulina sérica (μU/mL) | Numérica continua |
| mass | Índice de masa corporal (IMC) | Numérica continua |
| pedi | Función pediátrica de diabetes | Numérica continua |
| age  | Edad (años) | Entero |
| class | Variable objetivo (0 = no diabético, 1 = diabético) | Categórica binaria |


## 2. Necesidad del escalado

Las variables numéricas del dataset tienen **unidades y rangos muy diferentes**:

`plas` (glucosa) puede variar entre 50 y 200  
`insu` (insulina) puede llegar a valores superiores a 600  
`mass` (IMC) suele estar entre 15 y 50  
`pedi` (función pediátrica) toma valores pequeños (0.1–2.5)

Estos **diferentes órdenes de magnitud** afectan directamente a los algoritmos que calculan **distancias** (por ejemplo, KNN, SVM, PCA, K-Means) y se **regularizan** con penalizaciones L1/L2 (por ejemplo, regresión logística o redes neuronales).  

Sin escalado, las variables con valores grandes (como `insu`) **dominan el comportamiento del modelo**, reduciendo la importancia de otras variables más pequeñas (como `pedi`).

Por tanto, **es necesario escalar las variables numéricas** antes del entrenamiento.


## 3. Tipos de escalado aplicables

### a) **Normalización Min-Max**
Convierte los valores al rango [0, 1]:


x_escalado = (x - x_min)/(x_max - x_min)


**Ventajas:**
- Mantiene la forma de la distribución original.  
- Ideal para redes neuronales o métodos basados en distancia (KNN, SVM).

**Inconveniente:**
- Sensible a valores extremos (*outliers*).


### b) **Estandarización
Centra los datos en 0 y escala la desviación estándar a 1:

x_escalado = (x - u)/desv_tipica(x)


**Ventajas:**
- Menos afectado por las unidades originales.  
- Adecuado para la mayoría de los modelos lineales o de gradiente (Logistic Regression, SVM, PCA).  
- A menudo mejora la convergencia en modelos que usan optimización numérica.

**Inconveniente:**
- Los valores resultantes pueden ser negativos.  
- Puede ser menos intuitivo para interpretación directa.


```python

from sklearn.preprocessing import StandardScaler

# Copiar el dataset (sin la variable objetivo)
X = df.drop(columns=['class'])

# Escaladores
scaler_std = StandardScaler()


# Estandarización general
X_std = scaler_std.fit_transform(X)

X_std
```
